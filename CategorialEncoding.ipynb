```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

Label Encoding

```python
# Sample customer data
data = {
        'CustomerID': [1, 2, 3, 4, 5],
    'CustomerType': ['Regular', 'VIP', 'Regular', 'Premium', 'VIP']
}
customer_data = pd.DataFrame(data)

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'Education' column using label encoding
customer_data['CustomerType_encoded'] = label_encoder.fit_transform(customer_data['CustomerType'])

# Print the encoded data
customer_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Sample data: replace this with your actual data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'CustomerType': ['Regular', 'VIP', 'Regular', 'Premium', 'VIP']
}

# Creating a DataFrame from the sample data
df = pd.DataFrame(data)

# Ordinal mapping for CustomerType
customer_type_mapping = {
    'Regular': 0,
    'VIP': 1,
    'Premium': 2
}

# Applying label encoding using the mappings
df['EncodedCustomerType'] = df['CustomerType'].map(customer_type_mapping)

# Printing the transformed DataFrame
df.head()
```

One hot encoding

```python
import pandas as pd

# Sample data: replace this with your actual data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Country': ['USA', 'Canada', 'USA', 'UK', 'Thailand']
}

# Creating a DataFrame from the sample data
df = pd.DataFrame(data)

# Performing one-hot encoding on selected columns
encoded_df = pd.get_dummies(df, columns=['Country'], prefix=['Country'])
```
